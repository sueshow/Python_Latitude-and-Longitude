```python
#conda install -c conda-forge folium
import folium # 匯入 folium 套件
import pandas as pd 
# Folium預設使用OpenStreetMap
# 此外也可以使用如Stamen或Mapbox等地圖服務(Mapbox須申請並匯入API key)
```

```python
# (絕對路徑)
map_data = pd.read_csv('E:/Python/input/train.csv', header=None)
map_data.head()
```

```python
fmap = folium.Map((25.0133904,121.52245), tiles="Cartodb Positron",zoom_start=14)
fmap
```

```python
# 設定各個Marker的顏色，屬性與Icon類型
# 遊樂園
mark_amu = folium.Marker(location=[25.0143904,121.52445],
                             popup='<b>遊樂園<b>',
                             icon=folium.Icon(icon='map-marker', # Icon類型
                                              color='blue' # Marker顏色
                                             ))
# 將各Marker加入地圖。
fmap.add_child(child=mark_amu)
```

```python
# 設定各個Marker的顏色、屬性、Icon類型
# 學校
mark_sch = folium.Marker(location=[25.0133904,121.52245],
                             popup='<b>學校<b>',
                             icon=folium.Icon(icon='map-marker', # Icon類型
                                              color='red' # Marker顏色
                                             ))
# 將各Marker加入地圖。
fmap.add_child(child=mark_sch)
```

```python
# 設定各個Marker的顏色，屬性與Icon類型
# 社大
mark_sco = folium.Marker(location=[25.1133904,121.62245],
                             popup='<b>社大<b>',
                             icon=folium.Icon(icon='map-marker', # Icon類型
                                              color='orange' # Marker顏色
                                             ))
# 將各Marker加入地圖。
fmap.add_child(child=mark_sco)
```

```python
# 加入圓形標記
center_pos = [23.1133904,120.62245]
circle_pla = folium.Circle(location=center_pos,
                           color='green', # Circle 顏色
                           radius=3000, # Circle 寬度
                           popup='<b>站台<b>', # 彈出視窗內容
                           #fill=True, # 填滿中間區域
                           #fill_color='#3186cc',
                           fill_opacity=0.5 # 設定透明度
                          )
fmap.add_child(child=circle_pla)
```

```python
# Folium將繪製好的地圖匯出成HTML檔案
# 執行後會在Python檔案所在資料夾建立一個名為map.html的檔案，可直接用瀏覽器開啟觀看結果
fmap.save('map.html')
```
